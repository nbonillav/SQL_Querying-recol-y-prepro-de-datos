```python
from pandas_profiling import ProfileReport

# Mostrar todas las columnas del dataframe
pd.set_option('display.max_columns', None)
```

```python
# Importar archivo de excel
df = pd.read_excel('Lab_2_ Reporte de Producción_v2.xlsx')
df
```

## Puntos para el Data Cleansing
- Línea 2 encabezados
- Eliminar linea del Mix total
- Reemplazar los NaN  con las debidas categorías
- Eiminar Grand-total
- Trasponer las columnas año y mes(Febrero-2023)
- Eliminar columna Date


1. Promover Títulos

```python
# Asignar lista de títulos
titulos = df.iloc[2]
df.columns = titulos
df.reset_index(drop=True, inplace=True)
df.head(5)
```

2. Filtrar líneas

```python
# Elegir a partir de la línea 3
df_filtrado = df.iloc[3:]

#Eliminar filas de totales
df_sinTotales = df_filtrado[~ df_filtrado['Category'].str.contains('Total', na = False)]

#Remover nulos en columna de segment code
df_sinTotales2 = df_sinTotales.dropna(subset=['Segment Code'])

df_sinTotales.head(20)
```

```python
# Rellenar columna de Categoría
df_sinTotales2['Category'].fillna(method='ffill', inplace=True)
df_sinTotales2.head(6)
```

3. Transponer

```python
df_melt = pd.melt(df_sinTotales2, id_vars = df_sinTotales2.columns[:3], var_name='Date', value_name = 'Value')
print(df_melt.shape)
df_melt
```

4. Promover títulos

```python
# Remover Fechas que contienen total
df_melt_sinTot = df_melt[~df_melt['Date'].str.contains('Total', na = False)]
df_melt_sinTot.shape
df_melt_sinTot.head()
```

5. Separar Mes-Año

```python
# Separar lolumna de date
df_melt_sinTot['Year'] = df_melt_sinTot['Date'].str.split('-').str[1]
df_melt_sinTot['Month'] = df_melt_sinTot['Date'].str.split('-').str[0]
df_melt_sinTot.head()
```

6. Eliminación columna date

```python
# Remover columna Date
df_melt_sinTot = df_melt_sinTot.drop(columns=['Date'])
df_melt_sinTot.head()
```

## EDA with Profile Report
